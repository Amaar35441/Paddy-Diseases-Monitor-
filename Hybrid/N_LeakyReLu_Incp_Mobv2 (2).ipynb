```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
import matplotlib.pyplot as plt

# Define InceptionV3 Base Model
def inceptionv3_base(input_shape=(224, 224, 3)):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv2D(32, (3, 3), strides=(2, 2), padding='valid', activation=None)(inputs)
    x = layers.LeakyReLU(alpha=0.1)(x)
    x = layers.Conv2D(32, (3, 3), padding='valid', activation=None)(x)
    x = layers.LeakyReLU(alpha=0.1)(x)
    x = layers.Conv2D(64, (3, 3), padding='same', activation=None)(x)
    x = layers.LeakyReLU(alpha=0.1)(x)
    x = layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2), padding='valid')(x)
    x = layers.Conv2D(80, (1, 1), padding='valid', activation=None)(x)
    x = layers.LeakyReLU(alpha=0.1)(x)
    x = layers.Conv2D(192, (3, 3), padding='valid', activation=None)(x)
    x = layers.LeakyReLU(alpha=0.1)(x)
    x = layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2), padding='valid')(x)
    x = GlobalAveragePooling2D()(x)
    return Model(inputs, x)

# Define MobileNetV2 Base Model Without Pre-trained Weights
def mobilenetv2_base(input_shape=(224, 224, 3)):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv2D(32, (3, 3), strides=(2, 2), padding='same', activation=None)(inputs)
    x = layers.LeakyReLU(alpha=0.1)(x)
    x = layers.BatchNormalization()(x)
    for _ in range(3):  # Bottleneck blocks
        shortcut = x
        x = layers.Conv2D(32, (1, 1), padding='same', activation=None)(x)
        x = layers.LeakyReLU(alpha=0.1)(x)
        x = layers.BatchNormalization()(x)
        x = layers.DepthwiseConv2D((3, 3), padding='same', activation=None)(x)
        x = layers.LeakyReLU(alpha=0.1)(x)
        x = layers.BatchNormalization()(x)
        x = layers.Conv2D(32, (1, 1), padding='same', activation=None)(x)
        x = layers.LeakyReLU(alpha=0.1)(x)
        x = layers.BatchNormalization()(x)
        x = layers.Add()([x, shortcut])
    x = GlobalAveragePooling2D()(x)
    return Model(inputs, x)

# Build Hybrid Model
input_shape = (224, 224, 3)
inputs = layers.Input(shape=input_shape)

# InceptionV3 Base
inception_base = inceptionv3_base(input_shape)
inception_features = inception_base(inputs)

# MobileNetV2 Base
mobilenet_base = mobilenetv2_base(input_shape)
mobilenet_features = mobilenet_base(inputs)

# Combine Features
x = Concatenate()([inception_features, mobilenet_features])

# Fully Connected Layers
x = Dense(1024, activation='relu')(x)
outputs = Dense(5, activation='softmax')(x)  # Adjust for your number of classes

# Final Model
hybrid_model = Model(inputs, outputs)
hybrid_model.summary()

# Compile Model
hybrid_model.compile(optimizer=Adam(learning_rate=0.0001),
                     loss='sparse_categorical_crossentropy',
                     metrics=['accuracy'])


# Directory for the dataset
data_dir = '/content/rice-leaf-diseases-detection-'
train_datagen = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input)
val_datagen = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input)
test_datagen = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input)

train_generator = train_datagen.flow_from_directory(
    os.path.join(data_dir, 'train'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='sparse'
)

val_generator = val_datagen.flow_from_directory(
    os.path.join(data_dir, 'val'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='sparse'
)

test_generator = test_datagen.flow_from_directory(
    os.path.join(data_dir, 'test'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='sparse'
)

# Train the Hybrid Model
history = hybrid_model.fit(train_generator, epochs=70, validation_data=val_generator)

# Plot Training History
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Evaluate Model
def evaluate_model(generator, model, classes, title=None):
    if title:
        print(f"--- {title} ---")
    y_pred = np.argmax(model.predict(generator), axis=1)
    precision = precision_score(generator.labels, y_pred, average='weighted')
    recall = recall_score(generator.labels, y_pred, average='weighted')
    f1 = f1_score(generator.labels, y_pred, average='weighted')
    accuracy = accuracy_score(generator.labels, y_pred)
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1}")
    print(f"Accuracy: {accuracy}")

# Test the hybrid model on the test set
evaluate_model(test_generator, hybrid_model,
               ['bacterial_leaf_blight', 'brown_spot', 'healthy', 'leaf_blast', 'rice_hispa'],
               title="Performance on Test Data")
```
